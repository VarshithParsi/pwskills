Question 1: What is hypothesis testing in statistics?
Answer:
Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves formulating two hypotheses: the null hypothesis (H₀) and the alternative hypothesis (H₁), and then using sample data to determine whether to reject H₀. The process includes selecting a significance level, calculating a test statistic, and comparing it to a critical value or using a p-value to make a decision.


Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:
Null Hypothesis (H₀): Assumes no effect or no difference; it is the default or status quo assumption.

Alternative Hypothesis (H₁): Represents the claim or effect we are testing for; it suggests a difference or effect exists.

The null hypothesis is tested directly, and the alternative is accepted only if there is strong evidence against H₀.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer:
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05 or 0.01. It sets the threshold for the p-value:
If p-value ≤ α, reject H₀.
If p-value > α, fail to reject H₀.
It helps control the risk of making incorrect conclusions.

Question 4: What are Type I and Type II errors? Give examples of each.

Answer:

Type I Error (False Positive): Rejecting H₀ when it is true.
Example: Concluding a drug is effective when it actually isn’t.

Type II Error (False Negative): Failing to reject H₀ when H₁ is true.
Example: Concluding a drug is not effective when it actually is.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

Answer:
Z-test: Used when population variance is known and sample size is large (n > 30).
T-test: Used when population variance is unknown and sample size is small (n ≤ 30). Both tests compare sample means but differ in assumptions about the population.

```python
# 6) Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5
size = 1000

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 7) Implement hypothesis testing using Z-statistics for a sample dataset in Python
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu = 50  # population mean
sigma = 1  # population standard deviation
# Z-test
sample_mean = np.mean(sample_data)
n = len(sample_data)
z = (sample_mean - mu) / (sigma / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
# 8)Simulate data from a normal distribution and calculate the 95% confidence interval
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate data
data = np.random.normal(loc=100, scale=15, size=100)

# Confidence interval
mean = np.mean(data)
sem = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print(f"Mean: {mean:.2f}")
print(f"95% Confidence Interval: {ci}")

# Plot
plt.hist(data, bins=15, edgecolor='black')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 9) Calculate Z-scores and visualize standardized data
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = np.random.normal(100, 15, 100)

# Z-scores
mean = np.mean(data)
std = np.std(data)
z_scores = [(x - mean) / std for x in data]

# Plot
plt.hist(z_scores, bins=15, edgecolor='black')
plt.title('Z-scores Histogram')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print("Z-scores represent how many standard deviations a data point is from the mean.")
```
